```python
import os
import numpy as np
import pandas as pd

from scipy.special import softmax

from core.data.definitions import *
from core.data.io import *
```

#### Load Predictions on Validation Set

```python
EXPERIMENT_DIR = 'roberta-25-06-2020'

train, _ = load_datasets('in/', clean=True)
val_preds = pd.read_csv(EXPERIMENT_DIR+'/val_with_logits.csv')
```

```python
train.head()
```

```python
val_preds['Id'] = val_preds['Id'] - 1
val_preds.sort_values('Id', inplace=True)

df = train.iloc[val_preds['Id']]
```

```python
df.head()
```

```python
val_preds[['NegativeLogits', 'PositiveLogits']] = val_preds[['NegativeLogits', 
                                                             'PositiveLogits']].apply(lambda x: softmax(x), axis=1)
val_preds.head()
```

#### Check Accuracy of Predictions

```python
np.sum(df['y'].values == val_preds['Prediction'].values)/val_preds.shape[0]
```

#### Incorrect Predictions (Validation Set)

```python
LIMIT = None

c = 0
for i, t, p in zip(val_preds['Id'].values, df['y'].values, val_preds['Prediction'].values):
    if t != p:
        c+=1
        print ('=========================================================')
        print ('Tweet : ', df.loc[i]['text'].strip())
        print ('True Label : {}\t Predicted Label: {}'.format(t, p))
        col = val_preds[val_preds['Id'] == i]
        print ('+ve Logits: {}\t -ve Logits: {}\n'.format(col['PositiveLogits'].values[0], 
                                                          col['NegativeLogits'].values[0]))   
    if LIMIT and c > LIMIT:
        break
```
